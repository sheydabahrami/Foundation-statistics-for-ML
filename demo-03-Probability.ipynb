#### Import Libraries

```python
import random
import numpy as np

import scipy.stats

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

#### Small Samples

```python
def flip(n):

    result = []

    for i in range(1, n + 1):
        result.append(random.choice(["Heads", "Tails"]))

    return result
```

```python
def roll(n):
    
    result = []

    for i in range(1, n + 1):
        result.append(random.choice([1, 2, 3, 4, 5, 6]))
        
    return result
```

```python
result = roll(10)

result
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(x=result)

plt.xlabel('Dice Roll')
plt.title('Counts')

plt.show()
```

```python
result = roll(100)

plt.figure(figsize=(12, 8))

sns.countplot(x=result)

plt.xlabel('Dice Roll')
plt.title('Counts')

plt.show()
```
